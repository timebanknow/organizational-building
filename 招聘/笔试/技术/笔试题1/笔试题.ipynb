# **<center>扑克牌图像识别与博弈决策系统笔试题</center>**
&emsp;&emsp;本试题分为五个部分，涵盖图像生成、识别、对抗遮挡及策略推理。
## Part 1：规则化发牌与图像生成
### 目标
&emsp;&emsp;模拟真实扑克牌发牌及人手持牌的图像生成。人手持牌视角示例见图1。

<div align=center><img src="图片1.png"></div>

<center>图1 人手持牌视角示例</center>

### 要求
1.编写代码实现扑克牌洗牌与发牌逻辑。

2.假设玩法为“斗地主”或“简化版比大小”等。

3.为每位玩家生成手持牌图像，需满足：

- 扑克牌扇形展开；

- 相邻牌片部分重叠、牌面向上；

- 背景为绿色桌面，视角与人眼持牌一致。

4.将图像保存为.jpg格式，文件命名为player_1.jpg、player_2.jpg等。
### 提交内容
1..jpg格式的手持牌图像样例（每位玩家一张）。

2.包含发牌与图像生成逻辑的源代码（.py或.ipynb）。

3.简要说明文档（PDF），包含方法原理、流程图及关键实现代码或伪代码。
## Part 2：无遮挡识别扑克牌
### 目标
&emsp;&emsp;识别正常持牌视角图像中的所有扑克牌（花色+点数）。正常持牌视角示例见图2。

### 要求

- 模板匹配；

- 传统机器学习或深度学习分类器。

2.输入：单张手牌图像（如player_1.jpg），输出格式示例：

    Player 1: ['3♣', 'Q♠', 'A♦']

## Part 3：设计遮挡样式防机器识别
### 目标
&emsp;&emsp;模拟赌场赌客防摄像头偷窥的遮挡动作，评估对机器识别的干扰效果。赌场赌客防摄像头偷窥动作示例见图3。

<div align=center><img src="图片3.png"></div>

<center>图3 赌场赌客防摄像头偷窥动作示例</center>

### 要求
1.枚举并分析至少三种现实中可能的遮挡动作：

- 手指或手掌遮挡牌角；

- 其他道具（如纸张、衣袖）遮挡牌面；

- 身体姿态遮挡整个手牌。

2.编程实现上述遮挡效果，对已有手牌图像进行仿真处理。

3.对比并量化前后识别准确率，回答：

- 哪种遮挡方式对机器最具干扰性？

- 哪种方式对人眼仍能轻松识别？
### 提交内容
1.遮挡后手牌图像样例（至少3种方式，各3张）。

2.遮挡仿真源代码。

3.结果对比表及分析说明（PDF），含准确率变化及可视化示例等。
## Part 4：识别遮挡下的扑克牌
### 目标
&emsp;&emsp;在牌角或牌面被部分遮挡的条件下，探索提高机器识别鲁棒性的技术方案。
### 要求
1.结合上下文信息、颜色和几何特征等，提出识别补偿策略。

2.可选用深度学习方法（如YOLOv5/6、ResNet等）进行遮挡鲁棒识别。

3.输出遮挡环境下的识别准确率，并与无遮挡环境进行对比分析。
### 提交内容
1.遮挡环境下的新测试图像集（至少20张）。

2.识别改进代码及模型文件。

3.实验结果报告（PDF），包含准确率对比、错误分析与改进思路。
## Part 5：基于识别结果进行出牌策略推理
### 目标
&emsp;&emsp;在已知自己手牌和历史出牌记录的前提下，制定出牌策略以提升胜率。
### 要求
1.玩法与第一部分相同。

2.输出本轮应出的最佳牌组。

3.策略可包括：

- 保留大牌或炸弹；

- 优先出单张、对子或顺子；

- 根据对手已出牌预测对方剩余手牌。

4.请说明所用策略的核心逻辑等，并给出关键实现代码或伪代码。
### 提交内容
1.策略推理源代码。

2.多组测试用例及运行结果示例。

3.简要策略说明文档（PDF），阐述核心逻辑等。

**注意：** 可使用Python、OpenCV、PIL、PyTorch、TensorFlow等常用库。如果使用深度学习等需要自己寻找数据集。
